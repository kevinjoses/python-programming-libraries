# Pandas

```python
# Creating a dataframe using  list inside a dictionary
import pandas as pd
reg = {'Roll_No':[1,2,3,4,5],'Name':['Shahnavas','Emmanual','Kevin','Sajad','Kiran'],'Score':[88,77,98,66,55]}
df = pd.DataFrame(reg)
df
```

```python
# Creating a Dataframe using tuple inside a list

reg = [(1,'Shahnavas',66),(2,'Emmanual',77),(3,'Kevin',66)]
df = pd.DataFrame(reg,columns=['Roll_no','Name','Score'])
df
```

```python
# Creating a Dataframe using dictionary inside a list

reg = [{'Roll_No':1,'Name':'Shahnavas','Score':77},
       {'Roll_No':2,'Name':'Emmanual','Score':77},
       {'Roll_No':3,'Name':'Kevin','Score':77}]
df = pd.DataFrame(reg)
df
```
